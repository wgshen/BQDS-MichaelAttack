```python
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import trange
```

```python
evids_3 = np.load('./data/model_evidence_O1O2O3O4.npy')
evids_2 = np.load('./data/model_evidence_O1O2O3.npy')
evids_1 = np.load('./data/model_evidence_O1O2.npy')
```

```python
plt.figure(figsize=(20, 6))
x = np.arange(len(evids_1)) * 100
plt.plot(x, evids_3, label='O1O2O3O4')
plt.plot(x, evids_2, label='O1O2O3')
plt.plot(x, evids_1, label='O1O2')
plt.yscale('log')
plt.legend(fontsize=15)
plt.xlabel('n_sample', fontsize=20)
plt.ylabel('evid', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
evid_sum = evids_1[-1] + evids_2[-1] + evids_3[-1]
evids_norm = np.array([evids_1[-1], evids_2[-1], evids_3[-1]])
evids_norm /= evid_sum
plt.figure(figsize=(10, 4))
plt.bar(['O1O2', 'O1O2O3', 'O1O2O3O4'], evids_norm)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('P(model | data)', fontsize=20)
# plt.savefig('./model_compare_445nm_fixR_randomM.pdf', bbox_inches='tight')
plt.show()
```
